```python

manual_field_names = {
    "topmostSubform[0].Page1[0].f1_01[0]": "Tax year beginning", "topmostSubform[0].Page1[0].f1_02[0]": "Tax year ending","topmostSubform[0].Page1[0].f1_03[0]": "Year", "topmostSubform[0].Page1[0].f1_04[0]":"your first name and middle initial",
    "topmostSubform[0].Page1[0].f1_05[0]" : "last name", "topmostSubform[0].Page1[0].f1_06[0]" : "social security number", "topmostSubform[0].Page1[0].f1_07[0]":"spouse first name and middle initial", "topmostSubform[0].Page1[0].f1_08[0]" : "spouse last name",
    "topmostSubform[0].Page1[0].f1_09[0]" : "spouse social security number","topmostSubform[0].Page1[0].Address_ReadOrder[0].f1_10[0]" : "Home address (number and street)", "topmostSubform[0].Page1[0].Address_ReadOrder[0].f1_11[0]" : "Apt no",
    "topmostSubform[0].Page1[0].Address_ReadOrder[0].f1_12[0]" : "city, town or post office", "topmostSubform[0].Page1[0].Address_ReadOrder[0].f1_13[0]" : "state","topmostSubform[0].Page1[0].Address_ReadOrder[0].f1_14[0]" : "zip code", "topmostSubform[0].Page1[0].Address_ReadOrder[0].f1_15[0]" : "foreign country", 
    "topmostSubform[0].Page1[0].Address_ReadOrder[0].f1_16[0]" : "foreign state", "topmostSubform[0].Page1[0].Address_ReadOrder[0].f1_17[0]" : "pin code","topmostSubform[0].Page1[0].c1_1[0]" : "President Election Fund Donation", "topmostSubform[0].Page1[0].c1_2[0]" : "President Election Fund Donation Spouse",
    "topmostSubform[0].Page1[0].c1_3[0]" : "Filing Status : Single", "topmostSubform[0].Page1[0].c1_3[1]" : "Filing Status : Head of Household","topmostSubform[0].Page1[0].c1_3[2]" : "Filing Status : Married Filing Jointly", "topmostSubform[0].Page1[0].c1_3[3]" : "Filing Status: Married Filing Separately",
    "topmostSubform[0].Page1[0].c1_3[4]" : "Filing Status : Quaifying surviving spouse", "topmostSubform[0].Page1[0].f1_18[0]" : "Enter name if you selected any of previous three", "topmostSubform[0].Page1[0].c1_4[0]" : "Digital Assets : Yes", "topmostSubform[0].Page1[0].c1_4[1]" : "Digital assets : No",
    "topmostSubform[0].Page1[0].c1_5[0]" : "Standard Deduction : You as dependent",  "topmostSubform[0].Page1[0].c1_6[0]" : "Standard Deduction : Your spouse as dependent","topmostSubform[0].Page1[0].c1_7[0]" : "Standard Deduction : Separate return or dual-status",  "topmostSubform[0].Page1[0].c1_8[0]" : "Age/Blindness : You were born before 1959",
    "topmostSubform[0].Page1[0].c1_9[0]" : "Age/Blindness : You are blind", "topmostSubform[0].Page1[0].c1_10[0]" : "Age/Blindness : Spouse born before 1959", "topmostSubform[0].Page1[0].c1_11[0]" : "Age/Blindness : spouse is blind","topmostSubform[0].Page1[0].Dependents_ReadOrder[0].c1_12[0]" : "If more then 4 dependents",  
    "topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].f1_19[0]" : "Dependent1: First Name", "topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].f1_20[0]" : "Dependent1: SSN","topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].f1_21[0]" : "Dependent1: relationship to you", "topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].c1_13[0]" : "Dependent1 : Child tax credit","topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].c1_14[0]" : "Dependent1 : Credit for others",  
    "topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].f1_22[0]" : "Dependent2: First Name", "topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].f1_23[0]" : "Dependent2: SSN","topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].f1_24[0]" : "Dependent2: relationship to you", "topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].c1_15[0]" : "Dependent2 : Child tax credit",
    "topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].c1_16[0]" : "Dependent2 : Credit for others", "topmostSubform[0].Page1[0].Table_Dependents[0].Row3[0].f1_25[0]": "Dependent3: First Name", "topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].f1_26[0]" : "Dependent3: SSN","topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].f1_27[0]" : "Dependent3: relationship to you", 
    "topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].c1_17[0]" : "Dependent3 : Child tax credit","topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].c1_18[0]" : "Dependent3 : Credit for others","topmostSubform[0].Page1[0].Table_Dependents[0].Row3[0].f1_28[0]": "Dependent4: First Name","topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].f1_29[0]" : "Dependent4: SSN",
    "topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].f1_30[0]" : "Dependent4: relationship to you", "topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].c1_19[0]" : "Dependent4 : Child tax credit","topmostSubform[0].Page1[0].Table_Dependents[0].Row1[0].c1_20[0]" : "Dependent4 : Credit for others","topmostSubform[0].Page1[0].f1_31[0]" : "Total amount from Form(s) W-2, box 1",
    "topmostSubform[0].Page1[0].f1_32[0]" : "Household employee wages", "topmostSubform[0].Page1[0].f1_33[0]" : "Tip income","topmostSubform[0].Page1[0].f1_34[0]" : "Medicaid waiver payments","topmostSubform[0].Page1[0].f1_35[0]" : "Taxable dependent care benefits",  "topmostSubform[0].Page1[0].f1_36[0]" : "Employer-provided adoption benefits", "topmostSubform[0].Page1[0].f1_37[0]" : "Wages from Form 8919", 
    "topmostSubform[0].Page1[0].f1_38[0]" : "Other earned income", "topmostSubform[0].Page1[0].f1_39[0]" : "Nontaxable combat pay election","topmostSubform[0].Page1[0].f1_40[0]" : "Add lines 1a through 1h","topmostSubform[0].Page1[0].f1_41[0]" : "Tax-exempt interest","topmostSubform[0].Page1[0].f1_42[0]" : "Taxable interest", "topmostSubform[0].Page1[0].f1_43[0]" : "Qualified dividends",
    "topmostSubform[0].Page1[0].f1_44[0]" : "Ordinary dividends", "topmostSubform[0].Page1[0].Line4a-11_ReadOrder[0].f1_45[0]" : "IRA distributions", "topmostSubform[0].Page1[0].Line4a-11_ReadOrder[0].f1_46[0]" : "Taxable amount","topmostSubform[0].Page1[0].Line4a-11_ReadOrder[0].f1_47[0]" : "Pensions and annuities", "topmostSubform[0].Page1[0].Line4a-11_ReadOrder[0].f1_48[0]" : "Taxable amount", "topmostSubform[0].Page1[0].Line4a-11_ReadOrder[0].f1_49[0]" : "Social security benefits", 
    "topmostSubform[0].Page1[0].Line4a-11_ReadOrder[0].f1_50[0]" : "Taxable amount", "topmostSubform[0].Page1[0].Line4a-11_ReadOrder[0].c1_21[0]" : "If you elect to use the lump-sum election method","topmostSubform[0].Page1[0].Line4a-11_ReadOrder[0].c1_22[0]" : "Schedule D If not required","topmostSubform[0].Page1[0].Line4a-11_ReadOrder[0].f1_51[0]" : "Capital gain or (loss)","topmostSubform[0].Page1[0].Line4a-11_ReadOrder[0].f1_52[0]" : "Additional income from Schedule 1",
    "topmostSubform[0].Page1[0].Line4a-11_ReadOrder[0].f1_53[0]" : " total income", "topmostSubform[0].Page1[0].Line4a-11_ReadOrder[0].f1_54[0]" : "Adjustments to income from schedule 1","topmostSubform[0].Page1[0].Line4a-11_ReadOrder[0].f1_55[0]" : " adjusted gross income", "topmostSubform[0].Page1[0].f1_56[0]" : "Standard deduction", 
    "topmostSubform[0].Page1[0].f1_57[0]" : "Qualified business income deduction", "topmostSubform[0].Page1[0].f1_58[0]" : "Add lines 12 and 13", "topmostSubform[0].Page1[0].f1_59[0]" : "taxable income","topmostSubform[0].Page2[0].c2_1[0]" : "Tax Form 8814","topmostSubform[0].Page2[0].c2_2[0]" : "Tax Form 4972","topmostSubform[0].Page2[0].c2_3[0]" : "Other Tax Form",
    "topmostSubform[0].Page2[0].f2_01[0]" : "Tax","topmostSubform[0].Page2[0].f2_02[0]" : "Amount from Schedule 2","topmostSubform[0].Page2[0].f2_03[0]" : "Add lines 16 and 17", "topmostSubform[0].Page2[0].f2_04[0]" : "Child tax credit", "topmostSubform[0].Page2[0].f2_05[0]" : "Amount from schedule 3", "topmostSubform[0].Page2[0].f2_06[0]" : "Add lines 19 and 20","topmostSubform[0].Page2[0].f2_07[0]" : "Subtract line 21 from line 18", 
    "topmostSubform[0].Page2[0].f2_08[0]" : "Other taxes", "topmostSubform[0].Page2[0].f2_09[0]" : "Total Tax", "topmostSubform[0].Page2[0].f2_10[0]" : "Federal income tax withheld from", "topmostSubform[0].Page2[0].f2_11[0]" : "Form(s) W-2","topmostSubform[0].Page2[0].f2_12[0]" : "Form(s) 1099",     "topmostSubform[0].Page2[0].f2_13[0]" : "Other forms", "topmostSubform[0].Page2[0].f2_14[0]" : "Add lines 25a through 25c",
    "topmostSubform[0].Page2[0].f2_15[0]" : "2023 estimated tax payments","topmostSubform[0].Page2[0].f2_16[0]" : "Earned income credit","topmostSubform[0].Page2[0].f2_17[0]" : "Additional child tax credit", "topmostSubform[0].Page2[0].f2_18[0]" : "American opportunity credit", "topmostSubform[0].Page2[0].f2_19[0]" : "Reserved for future use", "topmostSubform[0].Page2[0].f2_20[0]" : "Amount from Schedule 3","topmostSubform[0].Page2[0].f2_21[0]" : "total other payments and refundable credits",
    "topmostSubform[0].Page2[0].f2_22[0]" : "total payments", "topmostSubform[0].Page2[0].f2_23[0]" : "overpaid","topmostSubform[0].Page2[0].c2_4[0]" : "If Form 8888 is attached","topmostSubform[0].Page2[0].f2_24[0]" : "refunded to you","topmostSubform[0].Page2[0].RoutingNo[0].f2_25[0]" : "Routing number","topmostSubform[0].Page2[0].c2_5[0]" : "Checking Type","topmostSubform[0].Page2[0].c2_5[1]" : "Savings Type", 
    "topmostSubform[0].Page2[0].AccountNo[0].f2_26[0]" : "Account number","topmostSubform[0].Page2[0].f2_27[0]" : "applied to your 2024 estimated tax","topmostSubform[0].Page2[0].f2_28[0]" : "amount you owe","topmostSubform[0].Page2[0].f2_29[0]" : " tax penalty","topmostSubform[0].Page2[0].c2_6[0]" : "Do you want to allow another person to discuss this return with the IRS : Yes",
    "topmostSubform[0].Page2[0].c2_6[1]" : "Do you want to allow another person to discuss this return with the IRS : No","topmostSubform[0].Page2[0].f2_30[0]" : "Designee’s name","topmostSubform[0].Page2[0].f2_31[0]" : "Phone no.","topmostSubform[0].Page2[0].f2_32[0]" : "PIN","topmostSubform[0].Page2[0].f2_33[0]" : "Your Occupation","topmostSubform[0].Page2[0].f2_34[0]" : "Your Identity PIN",
    "topmostSubform[0].Page2[0].f2_35[0]" : "Spouse Occupation","topmostSubform[0].Page2[0].f2_36[0]" : "Spouse Identity PIN","topmostSubform[0].Page2[0].f2_37[0]" : "Your phone no","topmostSubform[0].Page2[0].f2_38[0]" : "Your mail","topmostSubform[0].Page2[0].f2_39[0]" : "Paid Prepared Name","topmostSubform[0].Page2[0].f2_40[0]" : "PTIN","topmostSubform[0].Page2[0].c2_7[0]" : "Self Employed",
    "topmostSubform[0].Page2[0].f2_41[0]" : "Firm Name","topmostSubform[0].Page2[0].f2_42[0]" : "Firm Phone no", "topmostSubform[0].Page2[0].f2_43[0]" : "Firm Address", "topmostSubform[0].Page2[0].f2_44[0]" : "Firm EIN"
}

def extract_form_fields(pdf_path):
    
    form_fields_data = {}

    pdf_document = fitz.open(pdf_path)

    for page_number in range(len(pdf_document)):
        page = pdf_document.load_page(page_number)

        form_fields = page.widgets()

        for form_field in form_fields:
            field_name = form_field.field_name
            field_value = form_field.field_value

            if field_value == "Off" or not field_value.strip():
                continue
            
            if field_name in manual_field_names:
                annotated_name = manual_field_names[field_name]
                form_fields_data[field_name] = (annotated_name, field_value)

    return form_fields_data

form_fields_data = extract_form_fields("C:/Users/pawvaibh/Downloads/f1040.pdf")
for field_name, (annotated_name, value) in form_fields_data.items():
    print(f"{annotated_name}: {value}")
```
