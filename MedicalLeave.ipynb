```python
# Define the list of programs
programs = [
    'T-Mobile CC COL               ',
    'Airbnb CS COL                 ',
    'Auto Finance Collections      ',
    'Remitly CET IS BAQ            ',
    'GoDaddy Portuguese COLOMBIA   ',
    'AT&T CLG MOB Spanish - BQA    '
]

# Filter the DataFrame for the specified programs
top5_df = df[df['Program'].isin(programs)]

# Calculate the number of unique people under 'applied by'
unique_total_ppl = top5_df['Emp Id'].nunique()

total_ppl = len(top5_df['Emp Id'])

multiple = total_ppl - unique_total_ppl

# Calculate the number of unique people under 'Supervisor'
unique_supervisor = top5_df['Supervisor Name'].nunique()

# Calculate the number of unique people under 'applied by'
unique_applied_by = top5_df['Applied By'].nunique()

# Calculate the number of unique people under 'approved by'
unique_approved_by = top5_df['Approved By'].nunique()

# Calculate the average approval time for the "t-mobil" program
avg_approval_time_top5 = top5_df['Approval Time'].mean()

print("Number of people:", total_ppl)
print("Number of unique people:", unique_total_ppl)
print("Number of people who applied leave more than once:", multiple)
print("Number of unique people under 'Supervisor Name':", unique_supervisor)
print("Number of unique people under 'Applied By':", unique_applied_by)
print("Number of unique people under 'Approved By':", unique_approved_by)
print("Average approval time:", avg_approval_time_top5, "Days")
```

```python
# Create a dictionary to store the number of employees for each program
employees_in_program = {}

# Iterate over each program
for program in programs:
    # Count the unique employees for the current program
    num_employees = top5_df[top5_df['Program'] == program]['Emp Id'].nunique()
    # Store the result in the dictionary
    employees_in_program[program] = num_employees

# Print the number of employees in each program
print("Number of employees in each program:")
for program, num_employees in employees_in_program.items():
    print(f"{program}: {num_employees}")
```

```python
# Filter the top5_df to include only medical leaves
medical_leaves_top5 = top5_df[top5_df['Leave Type'] == 'Medical Leave']

# Count the number of times each employee has taken medical leave (top 10)
medical_leaves_count = medical_leaves_top5['Emp Name'].value_counts().head(10)

print("Top 10 Employees with the Most Medical Leaves:")
print(medical_leaves_count)
```

```python
# Assuming 'Leave Start Date' is a datetime column, you can extract the month-year
medical_leaves_top5['Month-Year'] = medical_leaves_top5['Leave Start Date'].dt.to_period('M')

# Group by 'Emp Name' and 'Month-Year', then count the number of medical leaves
medical_leaves_count_monthly = medical_leaves_top5.groupby(['Emp Name', 'Month-Year']).size()

# Find the top 10 employees with the highest medical leaves count
top10_employees = medical_leaves_count.index.tolist()

# Filter the monthly medical leaves count for the top 10 employees
top10_medical_leaves_monthly = medical_leaves_count_monthly.loc[top10_employees]

# Print the result
print("Medical Leaves Count for Top 10 Employees Month-wise:")
print(top10_medical_leaves_monthly)
```

```python

# Filter the DataFrame for medical leaves and for the top 10 employees
medical_leaves_top10 = top5_df[top5_df['Emp Name'].isin(medical_leaves_count.index)]

# Sort the DataFrame by 'Emp Name' and 'Leave Start Date'
medical_leaves_sorted = medical_leaves_top10.sort_values(by=['Emp Name', 'Leave Start Date'])

# Group by 'Emp Name' and calculate the difference between consecutive 'Leave Start Date'
medical_leaves_sorted['Days_Between_Leaves'] = medical_leaves_sorted.groupby('Emp Name')['Leave Start Date'].diff().dt.days

# Calculate the average number of days between leaves for each employee
avg_days_between_leaves_top10 = medical_leaves_sorted.groupby('Emp Name')['Days_Between_Leaves'].mean()

# Print the result
print("Average Number of Days Between Leaves for Top 10 Employees:")
print(avg_days_between_leaves_top10)
```

```python
# Group by 'Emp Name' and 'Program' and sum the total number of days/hours for medical leaves for each employee within each program
medical_leave_days_top5 = medical_leaves_top5.groupby(['Emp Name', 'Program'])['No. Of Days/Hours'].sum().reset_index()

# Sort the total leave days in descending order
sorted_medical_leave_days_top5 = medical_leave_days_top5.sort_values(by='No. Of Days/Hours', ascending=False)

# Get the top 5 employees with their corresponding programs
top5_medical_leave_employees = sorted_medical_leave_days_top5.head(5)
print('\n \n')
print(top5_medical_leave_employees[['Emp Name', 'Program', 'No. Of Days/Hours']])
```
