```python
import fitz 

def extract_form_fields(pdf_path):
    
    form_fields_data = {}

    pdf_document = fitz.open(pdf_path)

    for page_number in range(len(pdf_document)):
        page = pdf_document.load_page(page_number)

        form_fields = page.widgets()
       
        for form_field in form_fields:

            if form_field.field_value == form_field.on_state(): 
                field_name = form_field.field_name
                field_value = form_field.on_state()
                
                form_fields_data[field_name] = (field_name, field_value)
            else:
                field_name = form_field.field_name
                field_value = form_field.field_value
                form_fields_data[field_name] = (field_name, field_value)
            
    return form_fields_data

form_fields_data = extract_form_fields("C:/Users/pawvaibh/Downloads/1008irev.pdf")
for field_name, (name, value) in form_fields_data.items():
    print(f"{name}: {value}")
```
