```python
import fitz
from pdf2image import convert_from_path
import pytesseract
import cv2
import numpy as np

pytesseract.pytesseract.tesseract_cmd = "C:/Users/pawvaibh/AppData/Local/Programs/Tesseract-OCR/tesseract.exe"
def preprocess_image(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    denoised = cv2.medianBlur(binary, 3)
    contours, _ = cv2.findContours(denoised, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 1)
    return image

def extract_text_from_pdf(pdf_path):
    page_texts = []
    
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)
    
    for page_number in range(len(pdf_document)):
        page = pdf_document.load_page(page_number)
        text = page.get_text()
        
        if text:
            page_texts.append(text)
        else:
            # Convert the page to an image
            images = convert_from_path(pdf_path, first_page=page_number+1, last_page=page_number+1)
            cv_image = cv2.cvtColor(np.array(images[0]), cv2.COLOR_RGB2BGR)
            preprocessed_image = preprocess_image(cv_image)
            
            # Use pytesseract to extract text from the image
            image_text = pytesseract.image_to_string(preprocessed_image, lang='eng')
            page_texts.append(image_text)
    
    pdf_document.close()
    
    return page_texts

# Example usage
pdf_path = 'C:/Users/pawvaibh/Downloads/scansmpl1.pdf'
extracted_texts = extract_text_from_pdf(pdf_path)
for text in extracted_texts:
    print(text)
```

```python
import cv2
import numpy as np
import pdfplumber
import pytesseract
from pdf2image import convert_from_path


pytesseract.pytesseract.tesseract_cmd = "C:/Users/pawvaibh/AppData/Local/Programs/Tesseract-OCR/tesseract.exe"

def preprocess_image(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    denoised = cv2.medianBlur(binary, 3)
    contours, _ = cv2.findContours(denoised, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 1)
    return image

pdfpath = "C:/Users/pawvaibh/Downloads/scansmpl1.pdf"

page_texts = []

with pdfplumber.open(pdfpath) as pdf:
    for page_number in range(len(pdf.pages)):
        page = pdf.pages[page_number]
        text = page.extract_text()
        
        images = convert_from_path(pdfpath, first_page=page_number+1, last_page=page_number+1)
        cv_image = cv2.cvtColor(np.array(images[0]), cv2.COLOR_RGB2BGR)
                        
        height, width, _ = cv_image.shape
        if width > 2 * height:
            left_side = cv_image[:, 0:width // 2]
            right_side = cv_image[:, width // 2:]
           
            preprocessed_left = preprocess_image(left_side)
            preprocessed_right = preprocess_image(right_side)

            text_left = pytesseract.image_to_string(preprocessed_left, lang='eng')
            text_right = pytesseract.image_to_string(preprocessed_right, lang='eng')
                        
            page_texts.append([text_left, text_right])

        else:
                
            preprocessed_page = preprocess_image(cv_image)
               
            page_text = pytesseract.image_to_string(preprocessed_page, lang='eng')
                                
            page_texts.append(page_text)

for i, text in enumerate(page_texts):
    if isinstance(text, list):
        print(f"Page {i + 1} Text (Left Side):")
        print(text[0])
        print(f"Page {i + 1} Text (Right Side):")
        print(text[1])
        print("-------------")
    else:
        print(f"Page {i + 1} Text:")
        print(text)
        print("-------------")
```
