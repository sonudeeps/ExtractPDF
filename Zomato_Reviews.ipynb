```python
def extract_reviews_followers(metadata):
    parts = metadata.split(',')
    reviews = int(parts[0].strip().split()[0]) if 'Review' in parts[0] else 0
    followers = 0
    if len(parts) > 1:
        followers = int(parts[1].strip().split()[0]) if 'Followers' in parts[1] else 0
    return reviews, followers



# Apply the function to extract reviews and followers
df['Reviews'], df['Followers'] = zip(*df['Metadata'].apply(extract_reviews_followers))

# Display the DataFrame with extracted reviews and followers
print(df[['Restaurant', 'Reviewer', 'Rating', 'Reviews', 'Followers']])
```

```python
from stop_words import get_stop_words

# Get the English stopwords list
stopwords_list = get_stop_words('en')

# Add custom stopwords
custom_stopwords = ["one", "food"]

# Extend the stopwords list with custom stopwords
stopwords_list.extend(custom_stopwords)

# Print the list of stopwords
print(stopwords_list)
```

```python
import re

# Function to preprocess a single review
def preprocess_review(review):
    # Convert to lowercase
    review = review.lower()
    # Remove punctuation
    review = re.sub(r'[^\w\s]', '', review)
    # Tokenize the review
    tokens = review.split()
    # Remove stopwords
    filtered_tokens = [word for word in tokens if word not in stopwords_list]
    # Join the tokens back into a string
    preprocessed_review = ' '.join(filtered_tokens)
    return preprocessed_review

# Apply preprocessing to the 'Review' column
df['Preprocessed_Review'] = df['Review'].apply(preprocess_review)

# Print the preprocessed reviews
print(df['Preprocessed_Review'])
```

```python
from collections import Counter

# Tokenize the preprocessed reviews
tokenized_reviews = df['Preprocessed_Review'].str.split()

# Flatten the list of tokenized reviews
all_tokens = [token for review in tokenized_reviews for token in review]

# Calculate word frequencies
word_frequencies = Counter(all_tokens)

# Print the most common words and their frequencies
print("Top 10 Most Common Words:")
for word, frequency in word_frequencies.most_common(10):
    print(word, ":", frequency)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Initialize CountVectorizer to convert text to a bag of words
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')

# Fit and transform the preprocessed reviews
X = vectorizer.fit_transform(df['Preprocessed_Review'])

# Initialize LDA model
lda = LatentDirichletAllocation(n_components=5, random_state=42)

# Fit the LDA model to the transformed data
lda.fit(X)

# Print the top words for each topic
print("Top words for each topic:")
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(lda.components_):
    print(f"Topic {topic_idx+1}:")
    print([feature_names[i] for i in topic.argsort()[:-10 - 1:-1]])
```

```python
from textblob import TextBlob

# Function to get sentiment score
def get_sentiment_score(review):
    # Get the polarity score for the review
    polarity = TextBlob(review).sentiment.polarity
    # Determine the overall sentiment
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis to the 'Review' column
df['Sentiment'] = df['Preprocessed_Review'].apply(get_sentiment_score)

# Count the frequency of each sentiment
sentiment_counts = df['Sentiment'].value_counts()

# Print the frequency of each sentiment
print("Sentiment Frequency:")
print(sentiment_counts)

```

```python
# Group the DataFrame by 'Restaurant' and 'Sentiment' and count the occurrences
sentiment_counts_by_restaurant = df.groupby(['Restaurant', 'Sentiment']).size().unstack(fill_value=0)

# Find the restaurant with the highest number of positive, negative, and neutral sentiments
most_positive_restaurant = sentiment_counts_by_restaurant['Positive'].idxmax()
most_negative_restaurant = sentiment_counts_by_restaurant['Negative'].idxmax()
most_neutral_restaurant = sentiment_counts_by_restaurant['Neutral'].idxmax()

# Get the counts of positive, negative, and neutral sentiments for the most positive restaurant
positive_counts = sentiment_counts_by_restaurant.loc[most_positive_restaurant]

# Get the counts of positive, negative, and neutral sentiments for the most negative restaurant
negative_counts = sentiment_counts_by_restaurant.loc[most_negative_restaurant]

# Get the counts of positive, negative, and neutral sentiments for the most neutral restaurant
neutral_counts = sentiment_counts_by_restaurant.loc[most_neutral_restaurant]

# Print the results
print("Restaurant with the most positive sentiments:", most_positive_restaurant)
print("Positive Sentiments:", positive_counts['Positive'])
print("Negative Sentiments:", positive_counts['Negative'])
print("Neutral Sentiments:", positive_counts['Neutral'])

print("\nRestaurant with the most negative sentiments:", most_negative_restaurant)
print("Positive Sentiments:", negative_counts['Positive'])
print("Negative Sentiments:", negative_counts['Negative'])
print("Neutral Sentiments:", negative_counts['Neutral'])

print("\nRestaurant with the most neutral sentiments:", most_neutral_restaurant)
print("Positive Sentiments:", neutral_counts['Positive'])
print("Negative Sentiments:", neutral_counts['Negative'])
print("Neutral Sentiments:", neutral_counts['Neutral'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['Preprocessed_Review'], df['Sentiment'], test_size=0.2, random_state=42)

# Initialize TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=5000)

# Fit and transform the training data
X_train_vectorized = vectorizer.fit_transform(X_train)

# Initialize Logistic Regression model
classifier = LogisticRegression()

# Train the model
classifier.fit(X_train_vectorized, y_train)

# Transform the testing data
X_test_vectorized = vectorizer.transform(X_test)

# Predict sentiment for testing data
y_pred = classifier.predict(X_test_vectorized)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Group by reviewer and count the number of reviews
reviewer_counts = df.groupby('Reviewer').size()

# Filter reviewers with more than one review
multiple_reviews_reviewers = reviewer_counts[reviewer_counts > 9]

# Display reviewers with more than one review
if not multiple_reviews_reviewers.empty:
    print("Reviewers with more than one review:")
    print(multiple_reviews_reviewers)
else:
    print("No reviewers with more than one review found.")
```

```python
# Create an empty dictionary to store reviewer names and corresponding reviewed restaurants
reviewer_restaurants_dict = {}

# Iterate over the top reviewers with more than one review
for reviewer in multiple_reviews_reviewers.index:
    
    # Filter the DataFrame to include reviews by the current reviewer
    reviews_by_reviewer = df[df['Reviewer'] == reviewer]
    
    # Get the list of restaurants reviewed by the current reviewer
    reviewed_restaurants = reviews_by_reviewer['Restaurant'].unique()
    
    # Store the reviewer name and corresponding reviewed restaurants in the dictionary
    reviewer_restaurants_dict[reviewer] = reviewed_restaurants
    
    # Display the reviewer names and corresponding reviewed restaurants
    print("Reviewer names and corresponding reviewed restaurants:")
    for reviewer, restaurants in reviewer_restaurants_dict.items():
        print(f"Reviewer: {reviewer}")
        print("Reviewed Restaurants:")
        for restaurant in restaurants:
            print(restaurant)
        print()

```

```python
from collections import Counter

# Flatten the list of reviewed restaurants for all top 10 reviewers
all_reviewed_restaurants = [restaurant for restaurants_list in reviewer_restaurants_dict.values() for restaurant in restaurants_list]

# Count the occurrences of each restaurant
restaurant_counts = Counter(all_reviewed_restaurants)

# Find the most commonly reviewed restaurants
most_common_restaurants = restaurant_counts.most_common()

# Display the most commonly reviewed restaurants
if most_common_restaurants:
    print("Most commonly reviewed restaurants among the top 10 reviewers:")
    for restaurant, count in most_common_restaurants:
        print(f"{restaurant}: {count} reviews")
else:
    print("No restaurants reviewed by the top 10 reviewers.")
```
